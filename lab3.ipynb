```python
from scipy import optimize
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func(x):
    return (1/4)*x**4 - 2*x**3 + 4*x**2 - x + 3
```

```python
a = 1
b = 5
eps = 0.5
```

```python
x1 = a + (3-np.sqrt(5))*(b-a)/2
x2 = a + (np.sqrt(5)-1)*(b-a)/2
avgx = (a+b)/2
f1 = func(x1)
f2 = func(x2)
f = func(avgx)
arrxavg = np.array([avgx])
arrf = np.array([f])
while (b-a)>eps:
  if f1 <= f2:
    b = x2
    x2 = x1
    x1 = a + b - x1
  else:
    a = x1
    x1 = x2
    x2 = a + b - x2
  f1 = func(x1)
  f2 = func(x2)
```

```python
opt = optimize.minimize(func, x0=2.)
min_f = opt.fun
min_x1 = opt.x[0]
```

```python
min_x2 = (a+b)/2
min_fx = func(min_x2)

print("Minimum x:", min_x2)
print("Minimum value:", min_fx)
print("Loss = ", abs(min_x1-min_x2))
```

```python
xfunc = np.linspace(1, 5, num=100)
yfunc = func(xfunc)
plt.figure(figsize=(10, 6))
plt.plot(xfunc, yfunc, linewidth=2, label="f(x)")
plt.scatter(min_x2, min_fx, color='red', s=30, label="Minimum")
plt.scatter(min_x1, min_f, color='green', s=30, label="Real Minimum")
plt.axvline(min_x2, color='red', linestyle='--', alpha=0.5)
plt.axhline(min_fx, color='red', linestyle='--', alpha=0.5)
plt.title("Function and Minimum Point", fontsize=16)
plt.xlabel("x", fontsize=12)
plt.ylabel("f(x)", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.4)
plt.legend()
plt.show()
```
