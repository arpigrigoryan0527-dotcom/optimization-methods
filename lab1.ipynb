```python
from scipy import optimize
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func(x):
    return (1/4)*x**4 - 2*x**3 + 4*x**2 - x + 3
```

```python
a = 1
b = 5
eps = 0.5
n = 8
x = np.array([])
y = np.array([])
```

```python
for i in range(n+1):
    x = np.append(x,a+i*eps)
    y = np.append(y,func(a+i*eps))
```

```python
opt = optimize.minimize(func, x0=2.)
min_f = opt.fun
min_x1 = opt.x[0]
```

```python
min_index = np.argmin(y)

min_x2 = x[min_index]
min_fx = y[min_index]

print("Minimum x:", min_x2)
print("Minimum value:", min_fx)
print("Loss = ", abs(min_x1-min_x2))
```

```python
xfunc = np.linspace(a, b, num=100)
yfunc = func(xfunc)
plt.figure(figsize=(10, 6))
plt.plot(xfunc, yfunc, linewidth=2, label="f(x)")
plt.scatter(min_x2, min_fx, color='red', s=30, label="Minimum")
plt.scatter(min_x1, min_f, color='green', s=30, label="Real Minimum")
plt.axvline(min_x2, color='red', linestyle='--', alpha=0.5)
plt.axhline(min_fx, color='red', linestyle='--', alpha=0.5)
plt.title("Function and Minimum Point", fontsize=16)
plt.xlabel("x", fontsize=12)
plt.ylabel("f(x)", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.4)
plt.legend()
plt.show()
```
